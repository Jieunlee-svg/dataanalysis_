데이터 전처리

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
```

```python
df = pd.read_excel(filename)
df.head()
df.tail()
```

```python
c_col_name = df.columns[2]
df_공유받는사람 = df[df[c_col_name] == 1].copy()
df_공유하는사람 = df[df[c_col_name] == 2].copy()
```

```python
df_공유받는사람.to_excel("공유받는사람.xlsx", index=False)
df_공유하는사람.to_excel("공유하는사람.xlsx", index=False)
```

```python
files.download("공유받는사람.xlsx")
files.download("공유하는사람.xlsx")
```
