# 1-week retention

결론:
👥 분석 대상자 수 (가입 후 7일이 지난 사용자): 1061
🔁 1주 리텐션 유지 사용자 수 (가입일 다음날부터 7일 이내에 접속): 238
📊 1-week 리텐션율: 22.43%

가입일자와 마지막 접속일 사이의 차이가 1 이상 7 이하라는 것은→ 가입 후 1일차부터 7일차까지의 기간 중, 최소 한 번이라도 다시 접속한 경우를 의미

```python
import pandas as pd

# 기준일 설정
기준일자 = pd.to_datetime('2025-05-29')  # 또는 원하는 날짜

# 가입일자/마지막 접속일자 변환
df['가입일자'] = pd.to_datetime(df['가입일자'])
df['마지막 접속일'] = pd.to_datetime(df['마지막 접속일'])

# 분석 대상 사용자 필터링
df_분석대상 = df[df['가입일자'] <= (기준일자 - pd.Timedelta(days=7))].copy()

# 결과 확인
print(f"📊 분석 대상 사용자 수: {len(df_분석대상)}명")
display(df_분석대상[['가입일자', '마지막 접속일']].head())
```

```python
import pandas as pd

# 1. 날짜 형식 변환
df['가입일자'] = pd.to_datetime(df['가입일자'])
df['마지막 접속일'] = pd.to_datetime(df['마지막 접속일'])

# 2. 기준일자 정의 및 날짜만 추출
기준일자 = pd.to_datetime('2025-05-29')
기준일_날짜 = 기준일자.date()

# 3. 날짜만 추출 (시간 제거)
df['가입일자_날짜'] = df['가입일자'].dt.date
df['마지막 접속일_날짜'] = df['마지막 접속일'].dt.date

# 4. 조건에 맞는 사용자만 필터링하여 새 df 생성
df_1week분석대상 = df[df['가입일자_날짜'] <= (기준일_날짜 - pd.Timedelta(days=7))].copy()
```

```python
print(f"📊 분석 대상 사용자 수: {len(df_1week분석대상)}명")
display(df_1week분석대상.head())
```

```python
# 접속일수 계산(마지막 접속일과 가입일을 사이의 간격을 알아야 1일~7일 사이에 들어왔는지의 여부를 알 수 있기 때문)
df_1week분석대상['접속일수'] = (
    df_1week분석대상['마지막 접속일_날짜'] - df_1week분석대상['가입일자_날짜']
).apply(lambda x: x.days)
```

```python
#결과 확인
display(df_1week분석대상.head())
```

```python
#1-wek 리텐션 여부 계산(가입일 다음날부터 7일 이내 접속한 경우, 마지막 접속일과 가입일 사이의 간격이 1 이상 7 이하인 경우)
df_1week분석대상['1주리텐션'] = df_1week분석대상['접속일수'].between(1, 7)
```

```python
#결과 확인
print(df_1week분석대상['1주리텐션'].value_counts())
display(df_1week분석대상[['가입일자_날짜', '마지막 접속일_날짜', '접속일수', '1주리텐션']].head())
```

```python
# 리텐션 요약 통계
total_users = len(df_1week분석대상)
retained_users = df_1week분석대상['1주리텐션'].sum()
retention_rate = (retained_users / total_users) * 100 if total_users > 0 else 0

# 결과 출력
print(f"👥 분석 대상자 수 (가입 후 7일이 지난 사용자): {total_users}")
print(f"🔁 1주 리텐션 유지 사용자 수 (가입일 다음날부터 7일 이내에 접속): {retained_users}")
print(f"📊 1-week 리텐션율: {retention_rate:.2f}%")
```

```python
def define_age_group(age):
    if age < 20:
        return '10대'
    elif age < 30:
        return '20대'
    elif age < 40:
        return '30대'
    elif age < 50:
        return '40대'
    elif age < 60:
        return '50대'
    else:
        return '60대 이상'

# 나이를 기준으로 연령대 컬럼 추가
df_1week분석대상['연령대'] = df_1week분석대상['나이'].apply(define_age_group)
```
