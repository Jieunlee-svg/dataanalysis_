```python
import pandas as pd

# 1. ë‚ ì§œ í˜•ì‹ ë³€í™˜
df['ê°€ì…ì¼ì'] = pd.to_datetime(df['ê°€ì…ì¼ì'])
df['ë§ˆì§€ë§‰ ì ‘ì†ì¼'] = pd.to_datetime(df['ë§ˆì§€ë§‰ ì ‘ì†ì¼'])

# 2. ê¸°ì¤€ì¼ì ì •ì˜ ë° ë‚ ì§œë§Œ ì¶”ì¶œ
ê¸°ì¤€ì¼ì = pd.to_datetime('2025-05-29')
ê¸°ì¤€ì¼_ë‚ ì§œ = ê¸°ì¤€ì¼ì.date()

# 3. ë‚ ì§œë§Œ ì¶”ì¶œ (ì‹œê°„ ì œê±°)
df['ê°€ì…ì¼ì_ë‚ ì§œ'] = df['ê°€ì…ì¼ì'].dt.date
df['ë§ˆì§€ë§‰ ì ‘ì†ì¼_ë‚ ì§œ'] = df['ë§ˆì§€ë§‰ ì ‘ì†ì¼'].dt.date

# 4. ì¡°ê±´ì— ë§ëŠ” ì‚¬ìš©ìë§Œ í•„í„°ë§í•˜ì—¬ ìƒˆ df ìƒì„±
df_1weekë¶„ì„ëŒ€ìƒ = df[df['ê°€ì…ì¼ì_ë‚ ì§œ'] <= (ê¸°ì¤€ì¼_ë‚ ì§œ - pd.Timedelta(days=7))].copy()
```

```python
print(f"ğŸ“Š ë¶„ì„ ëŒ€ìƒ ì‚¬ìš©ì ìˆ˜: {len(df_1weekë¶„ì„ëŒ€ìƒ)}ëª…")
display(df_1weekë¶„ì„ëŒ€ìƒ.head())
```

```python
# ì ‘ì†ì¼ìˆ˜ ê³„ì‚°(ë§ˆì§€ë§‰ ì ‘ì†ì¼ê³¼ ê°€ì…ì¼ì„ ì‚¬ì´ì˜ ê°„ê²©ì„ ì•Œì•„ì•¼ 1ì¼~7ì¼ ì‚¬ì´ì— ë“¤ì–´ì™”ëŠ”ì§€ì˜ ì—¬ë¶€ë¥¼ ì•Œ ìˆ˜ ìˆê¸° ë•Œë¬¸)
df_1weekë¶„ì„ëŒ€ìƒ['ì ‘ì†ì¼ìˆ˜'] = (
    df_1weekë¶„ì„ëŒ€ìƒ['ë§ˆì§€ë§‰ ì ‘ì†ì¼_ë‚ ì§œ'] - df_1weekë¶„ì„ëŒ€ìƒ['ê°€ì…ì¼ì_ë‚ ì§œ']
).apply(lambda x: x.days)
```

```python
#ê²°ê³¼ í™•ì¸
display(df_1weekë¶„ì„ëŒ€ìƒ.head())
```

```python
#1-wek ë¦¬í…ì…˜ ì—¬ë¶€ ê³„ì‚°(ê°€ì…ì¼ ë‹¤ìŒë‚ ë¶€í„° 7ì¼ ì´ë‚´ ì ‘ì†í•œ ê²½ìš°, ë§ˆì§€ë§‰ ì ‘ì†ì¼ê³¼ ê°€ì…ì¼ ì‚¬ì´ì˜ ê°„ê²©ì´ 1 ì´ìƒ 7 ì´í•˜ì¸ ê²½ìš°)
df_1weekë¶„ì„ëŒ€ìƒ['1ì£¼ë¦¬í…ì…˜'] = df_1weekë¶„ì„ëŒ€ìƒ['ì ‘ì†ì¼ìˆ˜'].between(1, 7)
```

```python
#ê²°ê³¼ í™•ì¸
print(df_1weekë¶„ì„ëŒ€ìƒ['1ì£¼ë¦¬í…ì…˜'].value_counts())
display(df_1weekë¶„ì„ëŒ€ìƒ[['ê°€ì…ì¼ì_ë‚ ì§œ', 'ë§ˆì§€ë§‰ ì ‘ì†ì¼_ë‚ ì§œ', 'ì ‘ì†ì¼ìˆ˜', '1ì£¼ë¦¬í…ì…˜']].head())
```

```python
# ë¦¬í…ì…˜ ìš”ì•½ í†µê³„
total_users = len(df_1weekë¶„ì„ëŒ€ìƒ)
retained_users = df_1weekë¶„ì„ëŒ€ìƒ['1ì£¼ë¦¬í…ì…˜'].sum()
retention_rate = (retained_users / total_users) * 100 if total_users > 0 else 0

# ê²°ê³¼ ì¶œë ¥
print(f"ğŸ‘¥ ë¶„ì„ ëŒ€ìƒì ìˆ˜ (ê°€ì… í›„ 7ì¼ì´ ì§€ë‚œ ì‚¬ìš©ì): {total_users}")
print(f"ğŸ” 1ì£¼ ë¦¬í…ì…˜ ìœ ì§€ ì‚¬ìš©ì ìˆ˜ (ê°€ì…ì¼ ë‹¤ìŒë‚ ë¶€í„° 7ì¼ ì´ë‚´ì— ì ‘ì†): {retained_users}")
print(f"ğŸ“Š 1-week ë¦¬í…ì…˜ìœ¨: {retention_rate:.2f}%")
```
